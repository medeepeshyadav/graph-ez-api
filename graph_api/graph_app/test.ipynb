```python
import numpy as np
from multiprocessing import Pool
```

```python
def parallelize_apply(data, func, n_jobs):
    df_split = np.array_split(data, n_jobs)
    pool = Pool(n_jobs)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
```
